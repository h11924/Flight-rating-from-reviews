clearing the data

```python
#clearing airline.csv and storing the updated data in the same file
import pandas as pd

# Step 1: Read the CSV file and name the DataFrame
df = pd.read_csv('/content/airline.csv')

# Step 2: Drop the specified columns that are almost empty
columns_to_drop = ['ground_service_rating', 'wifi_connectivity_rating', 'value_money_rating', 'route']
df.drop(columns=columns_to_drop, inplace=True)

# Step 3: Calculate the average ratings for the specified columns while ignoring NaN values
average_columns = ['overall_rating', 'seat_comfort_rating', 'cabin_staff_rating', 'food_beverages_rating', 'inflight_entertainment_rating']
df['average_rating'] = df[average_columns].mean(axis=1, skipna=True)

# Step 4: Fill NaN values in the specified columns with the row average
for col in average_columns:
    df[col].fillna(df['average_rating'], inplace=True)

# Step 5: Replace empty 'author_country' values with 'unknown'
df['author_country'].fillna('unknown', inplace=True)

# Step 6: Group by airline name and calculate the average rating of all parameters
average_by_airline = df.groupby('airline_name')['average_rating'].mean().reset_index()

# Step 7: Display the average ratings by airline
print(average_by_airline)

# Step 8: Update the existing CSV file with the cleaned DataFrame
df.to_csv('/content/airline.csv', index=False)
```

```python
#seat_df contains the ratings and traveler types for airlines, here's the code for each probability-based analysis.
#probability distributing of the ratings
import pandas as pd

# Example data for demonstration (in actual usage, seat_df will be your DataFrame)
# Ratings (overall_rating) and traveler types (type_traveller)

# For the probability distribution of ratings
rating_counts = seat_df['overall_rating'].value_counts()

# Calculate the probability distribution of the ratings
rating_probabilities = rating_counts / rating_counts.sum()

# Display the probability distribution of ratings
print("Probability Distribution of Ratings:")
print(rating_probabilities)

# For conditional probability: Probability of receiving a high rating (9 or 10) for business travelers
high_ratings_business = seat_df[(seat_df['type_traveller'] == 'Business') & (seat_df['overall_rating'] >= 9)]
total_business_travelers = seat_df[seat_df['type_traveller'] == 'Business']

probability_high_rating_given_business = len(high_ratings_business) / len(total_business_travelers)
print(f"\nProbability of a high rating (9 or 10) given that the traveler is a business traveler: {probability_high_rating_given_business}")

# For expected value of the ratings
expected_value = sum(rating_probabilities.index * rating_probabilities)
print(f"\nExpected Value of Airline Ratings: {expected_value}")

# For the probability of receiving a rating greater than 8
high_ratings = seat_df[seat_df['overall_rating'] > 8]
total_ratings = seat_df['overall_rating'].count()

probability_high_rating = len(high_ratings) / total_ratings
print(f"\nProbability of receiving a rating above 8: {probability_high_rating}")

```

```python
#getting a list of the airports based on their lounge ratings
import pandas as pd

# Step 1: Read the updated CSV files
lounge_df = pd.read_csv('lounge.csv')
airport_df = pd.read_csv('airport.csv')

# Step 2: Extract unique airport names and their ratings
# Create a dictionary to hold airport names and their total ratings
airport_ratings_dict = {}

# Gather ratings from lounges
for index, row in lounge_df.iterrows():
    airport_name = row['airport']
    overall_rating = row['overall_rating']

    if airport_name and pd.notna(overall_rating):  # Ensure both are not null
        if airport_name not in airport_ratings_dict:
            airport_ratings_dict[airport_name] = []
        airport_ratings_dict[airport_name].append(overall_rating)

# Gather ratings from airports
for index, row in airport_df.iterrows():
    airport_name = row['airport_name']
    overall_rating = row['overall_rating']

    if airport_name and pd.notna(overall_rating):  # Ensure both are not null
        if airport_name not in airport_ratings_dict:
            airport_ratings_dict[airport_name] = []
        airport_ratings_dict[airport_name].append(overall_rating)

# Step 3: Calculate average ratings for each airport
average_airport_ratings = {}
for airport, ratings in airport_ratings_dict.items():
    average_airport_ratings[airport] = sum(ratings) / len(ratings)  # Calculate average

# Step 4: Sort the airports by average ratings in decreasing order
sorted_airports = sorted(average_airport_ratings.items(), key=lambda x: x[1], reverse=True)

# Step 5: Save the list of airports and their ratings in a data structure
airport_rating_list = [(airport, rating) for airport, rating in sorted_airports]

# Display the sorted airport ratings
print("List of Airports and Their Average Ratings:")
for airport, rating in airport_rating_list:
    print(f"{airport}: {rating}")

# Optional: Save the list to a CSV file
pd.DataFrame(airport_rating_list, columns=['Airport Name', 'Average Rating']).to_csv('sorted_airports_by_rating.csv', index=False)
```

```python
# ### 4. Likelihood Estimation ###
# Function to calculate likelihood for a category based on review text and keywords.
def calculate_likelihood(review_text, category_keywords):
    """Calculates the weighted likelihood score for a given review and category."""
    positive_score = sum([category_keywords['weights']['positive']
                          for word in category_keywords['positive'] if word in review_text.lower()])
    negative_score = sum([category_keywords['weights']['negative']
                          for word in category_keywords['negative'] if word in review_text.lower()])

    total_score = positive_score - negative_score
    return total_score
```

```python
# ### 7. Prediction ###
# Function to predict ratings for each category based on keyword likelihoods
# It predicts ratings for all categories by calculating the likelihood based on the presence of keywords
# The overall rating will be adjusted to be out of 10
def predict_ratings(review):
    """Predicts ratings for each category based on keyword likelihoods."""
    scores = {category: calculate_likelihood(review, keywords[category]) for category in keywords.keys()}
    return [
        max(1, min(10, round(scores['overall'] * 2))),   # Overall rating out of 10
        max(1, min(5, round(scores['queuing']))),        # Queuing
        max(1, min(5, round(scores['seat_comfort']))),   # Seat Comfort
        max(1, min(5, round(scores['cabin']))),          # Cabin
        max(1, min(5, round(scores['staff']))),          # Staff
        max(1, min(5, round(scores['value_for_money']))),# Value for Money
        max(1, min(5, round(scores['food']))),           # Food
        max(1, min(5, round(scores['shopping']))),       # Shopping
        max(1, min(5, round(scores['wifi']))),           # Wi-Fi
        max(1, min(5, round(scores['hygiene']))),        # Hygiene
        max(1, min(5, round(scores['entertainment']))),  # Entertainment
    ]
```

this uses the bayers intefreence to calculate the probabilities ,it normalises the probabilities and returns them

```python
 ### 6. Bayesian Inference ###
# Function to calculate probabilities for each rating based on review content
# Using the calculated likelihoods to estimate probabilities for each rating value
def calculate_probabilities(review):
    """Calculates probabilities for each rating (1 to 5 or 1 to 10 for overall) based on keyword likelihoods."""
    probabilities = {}
    for category in keywords.keys():
        # Get the total score for the category
        score = calculate_likelihood(review, keywords[category])
        # Convert score to probabilities for each rating from 1 to 5 (or 1 to 10 for overall)
        if category == 'overall':
            probabilities[category] = np.exp(score * 2) / np.sum([np.exp(score * 2) for i in range(1, 11)])
        else:
            probabilities[category] = np.exp(score) / np.sum([np.exp(score) for i in range(1, 6)])
    return probabilities
```

Define a Function to Calculate Probabilities

```python
def calculate_probabilities(preferences, category_keywords):
    """Calculates the probability of each rating (positive or negative) based on user preferences."""
    score = 0
    for keyword in preferences:
        if keyword in category_keywords["positive"]:
            score += 1  # Increase score for positive keywords
        elif keyword in category_keywords["negative"]:
            score -= 1  # Decrease score for negative keywords

    # Calculate probability using an exponential model
    positive_probability = np.exp(score) / (np.exp(score) + np.exp(-score))
    negative_probability = 1 - positive_probability
    return positive_probability, negative_probability
```

Predict the Best Options for the Itinerary Using Probabilities

```python
def predict_itinerary_with_probabilities(preferences):
    """Predicts an optimal itinerary based on user preferences, using probabilities."""

    # Calculate probabilities for each category
    transport_prob = calculate_probabilities(preferences['transport'], keywords['transport'])
    accommodation_prob = calculate_probabilities(preferences['accommodation'], keywords['accommodation'])
    food_prob = calculate_probabilities(preferences['food'], keywords['food'])
    activities_prob = calculate_probabilities(preferences['activities'], keywords['activities'])

    # Create the itinerary description based on the probabilities
    itinerary_description = f"Based on your preferences, here's an optimal itinerary:\n"

    # Transport
    if transport_prob[0] > 0.5:
        itinerary_description += f"Transport: Choose a fast, comfortable, and affordable option with a probability of {transport_prob[0]:.2f}.\n"
    else:
        itinerary_description += f"Transport: Expect delays or inconveniences with a probability of {transport_prob[1]:.2f}.\n"

    # Accommodation
    if accommodation_prob[0] > 0.5:
        itinerary_description += f"Accommodation: Stay at a clean, spacious, and comfortable place with a probability of {accommodation_prob[0]:.2f}.\n"
    else:
        itinerary_description += f"Accommodation: Consider budget options with possible noise or smaller rooms (probability: {accommodation_prob[1]:.2f}).\n"

    # Food
    if food_prob[0] > 0.5:
        itinerary_description += f"Food: Enjoy delicious, fresh meals with a probability of {food_prob[0]:.2f}.\n"
    else:
        itinerary_description += f"Food: Limited variety or quality with a probability of {food_prob[1]:.2f}.\n"

    # Activities
    if activities_prob[0] > 0.5:
        itinerary_description += f"Activities: Fun, exciting, and culturally enriching activities with a probability of {activities_prob[0]:.2f}.\n"
    else:
        itinerary_description += f"Activities: Expect more relaxed or budget-friendly options (probability: {activities_prob[1]:.2f}).\n"

    return itinerary_description
```
